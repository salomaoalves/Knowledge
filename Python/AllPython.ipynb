## Email

**Creating a Gmail token for authentication:**

1. Go to: https://security.google.com/settings/security/apppasswords
2. In 'Selecionar Dispositivo', select 'Others' and choose a name
3. Choose a name - the application name
4. Hit 'Gerar' and the token will be genareted

```python
def send_mail(body_text, subject, receiver_email):
  import smtplib
  import ssl
  from email.mime.text import MIMEText
  from email.mime.multipart import MIMEMultipart

  port = 465
  password = "bmtkjpqbhbzcnvdm"
  sender_email = "jjecarlos5@gmail.com"

  message = MIMEMultipart("alternative")
  message["Subject"] = subject
  message["From"] = sender_email
  message['To'] = ", ".join(receiver_email)

  part1 = MIMEText(body_text, "text") #other formats - like "html"

  # add html/plain text parts to MIMEMultipart message
  # the email client will try to render the last part first
  message.attach(part1)

  context = ssl.create_default_context()
  with smtplib.SMTP_SSL("smtp.gmail.com", port, context=context) as server:
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, message.as_string())

send_mail('NÃ£o houve relatorio!', 'Test', ['salomao.alves222@gmail.com'])
```

## SMS

Need to create a account in Twilio. To do, go here: https://www.twilio.com/blog/como-enviar-sms-python

```python
import os
from twilio.rest import Client

#Twilio authentication
account_sid = os.environ.get('TWILIO_ACCOUNT_SID')
auth_token = os.environ.get('TWILIO_AUTH_TOKEN')

client = Client(account_sid, auth_token)

client.messages.create(from_=os.environ.get('TWILIO_PHONE_NUMBER'), #or a string
                      to=os.environ.get('CELL_PHONE_NUMBER'),       # with the numb
                      body='SMS body message!')
```
