## Dimensionality Reduction

```julia
using VegaDatasets
using DataFrames
using Statistics
using MultivariateStats
using Plots

C = DataFrame(VegaDatasets.dataset("cars"))
dropmissing!(C)
M = Matrix(C[:,2:7]);
```

```julia
# PCA

data = (M .- mean(M,dims = 1))./ std(M,dims=1) # center and normalize the data

p = fit(PCA,data',maxoutdim=2) # fit PCA model
P = projection(p) # to obatain the projection matrix

firstRow = P'*(data[1,:]-mean(p)) # to transform a row with the new qty of cols
allData = MultivariateStats.transform(p, data') # notice that allData[:,1] is the same as P'*(data[1,:]-mean(p))

data2 = reconstruct(p, allData); # reconstruct testing observations (approximately)

# in allData, the cols are the records and the rows are the new cols (==maxoutdim)
Plots.scatter(allData[1,:],allData[2,:])
```

## AUC n Confusion Matrix

```julia
gt = [1, 1, 1, 1, 1, 1, 1, 2]
pred = [1, 1, 2, 2, 1, 1, 1, 1]
C = confusmat(2, gt, pred)   # compute confusion matrix
C ./ sum(C, dims=2)   # normalize per class
sum(diag(C)) / length(gt)  # compute correct rate from confusion matrix
correctrate(gt, pred)
C = confusmat(2, gt, pred)
```

## Clustering

```julia
using CSV
using Clustering
using Distances
using DataFrames

houses = CSV.read("data/newhouses.csv", DataFrame);
X = houses[!, [:latitude,:longitude]]
xmatrix = Matrix(X)'
```

```julia
# KMeans
C = kmeans(Matrix(X)', 10)
insertcols!(houses,3,:cluster10=>C.assignments);
```

```julia
# build a Distance Matrix
D = pairwise(Euclidean(), xmatrix, xmatrix,dims=2)
D2 = pairwise(SqEuclidean(), xmatrix,dims=2);
```

```julia
# DBSCAN
L = dbscan(D2, 0.05, 10)
insertcols!(houses,3,:dbscanclusters3=>L.assignments);
```

```julia
# kMedoid
K = kmedoids(D,10)
insertcols!(houses,3,:medoids_clusters=>K.assignments)
```

```julia
# Hierarchial
K = hclust(D)
L = cutree(K;k=10)
insertcols!(houses,3,:hclust_clusters=>L)
```

## Classification/Regression

```julia
using RDatasets
using GLMNet
using MLBase
using Random
using DecisionTree
using NearestNeighbors
using LIBSVM
using DataStructures
using StatsModels
using GLM
using Plots

iris = dataset("datasets", "iris")
X = Matrix(iris[:,1:4])
y = labelencode(labelmap(iris[:,5]), iris[:,5]);
```

```julia
# Useful functions
findaccuracy(predictedvals,groundtruthvals) = sum(predictedvals.==groundtruthvals)/length(groundtruthvals)
assign_class(predictedvalue) = argmin(abs.(predictedvalue .- [1,2,3]))
function perclass_splits(y,at)
    uids = unique(y)
    keepids = []
    for ui in uids
        curids = findall(y.==ui)
        rowids = randsubseq(curids, at)
        push!(keepids,rowids...)
    end
    return keepids
end

# split data
trainids = perclass_splits(y,0.7)
testids = setdiff(1:length(y),trainids);
```

```julia
# Linear Reg

xdata = repeat(1:0.5:10, inner=2)
ydata = 3 .+ xdata .+ 2 .* rand(length(xdata)) .-1

data = DataFrame(X=xdata, Y=ydata)
ols = lm(@formula(Y ~ X), data)

plot!(xdata,GLM.predict(ols));
```

```julia
# Logistic Reg

cats = dataset("MASS", "cats")

lmap = labelmap(cats[!,:Sex])
ci = labelencode(lmap, cats[!,:Sex])
data = DataFrame(X=cats[!,:HWt], Y=ci.-1)

probit = glm(@formula(Y ~ X), data, Binomial(), LogitLink())

scatter(data[!,:X],data[!,:Y],label="ground truth gender",color=6)
scatter!(data[!,:X],GLM.predict(probit),label="predicted gender",color=7);
```

```julia
# Lasso

path = glmnet(X[trainids,:], y[trainids])
cv = glmnetcv(X[trainids,:], y[trainids])
mylambda = path.lambda[argmin(cv.meanloss)] # best lambda
path = glmnet(X[trainids,:], y[trainids],lambda=[mylambda])

predictions_lasso = GLMNet.predict(path,X[testids,:])
assign_lasso = assign_class.(predictions_lasso)
findaccuracy(assign_lasso,y[testids]);
```

```julia
# Ridge

path = glmnet(X[trainids,:], y[trainids],alpha=0);
cv = glmnetcv(X[trainids,:], y[trainids],alpha=0)
mylambda = path.lambda[argmin(cv.meanloss)] # best lambda
path = glmnet(X[trainids,:], y[trainids],alpha=0,lambda=[mylambda]);

predictions_ridge = GLMNet.predict(path,X[testids,:])
predictions_ridge = assign_class.(predictions_ridge)
findaccuracy(predictions_ridge,y[testids])
```

```julia
# Elastic Net

path = glmnet(X[trainids,:], y[trainids],alpha=0.5);
cv = glmnetcv(X[trainids,:], y[trainids],alpha=0.5)
mylambda = path.lambda[argmin(cv.meanloss)]
path = glmnet(X[trainids,:], y[trainids],alpha=0.5,lambda=[mylambda]);

predictions_EN = GLMNet.predict(path,X[testids,:])
predictions_EN = assign_class.(predictions_EN)
findaccuracy(predictions_EN,y[testids])
```

```julia
# Decision Trees

odel = DecisionTreeClassifier(max_depth=2)
DecisionTree.fit!(model, X[trainids,:], y[trainids])

predictions_DT = DecisionTree.predict(model, X[testids,:])
findaccuracy(predictions_DT,y[testids])
```

```julia
# Random Forests

model = RandomForestClassifier(n_trees=20)
DecisionTree.fit!(model, X[trainids,:], y[trainids])

predictions_RF = DecisionTree.predict(model, X[testids,:])
findaccuracy(predictions_RF,y[testids])
```

```julia
# Nearest Neughbor

Xtrain = X[trainids,:]
ytrain = y[trainids]
kdtree = KDTree(Xtrain')
idxs, dists = knn(kdtree, X[testids,:]', 5, true)

c = ytrain[hcat(idxs...)]
possible_labels = map(i->counter(c[:,i]),1:size(c,2))

predictions_NN = map(i->parse(Int,string(string(argmax(possible_labels[i])))),1:size(c,2))
findaccuracy(predictions_NN,y[testids])
```

```julia
# Support Vector Machines

Xtrain = X[trainids,:]
ytrain = y[trainids]
model = svmtrain(Xtrain', ytrain)

predictions_SVM, decision_values = svmpredict(model, X[testids,:]')
findaccuracy(predictions_SVM,y[testids])
```
